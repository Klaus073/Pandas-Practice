# Loading Data

```python
census = pd.read_csv("acs2017_county_data.csv" , index_col = 1)
census.head(10)
```

# Dropping Columns

```python
census.drop('CountyId', axis=1, inplace=True)
```

# Resetting Index

```python
census = census.reset_index()
```

# Group By "State"

```python
census_grp = census.groupby('State')
```

```python
def print_groups(group_object):
    for name , group in group_object:
        print(name)
        print(group)
```

```python
print_groups(census_grp)
```

```python
census_grp.get_group('Texas')
```

## Aggregating:

```python
mig_112.agg(np.mean)
# mig_112 has 24 groups as seen above. This command will create mean for each group so in total we should have 24 means.
# The output has the identically structured indices as the original data, mi. As the original data, mi, the following
# output of aggregation has three indices (Symbol, Year, and Month).
```

```python
type(mig_112.agg(np.mean))
```

```python
mig_112.agg(np.mean).loc['AAPL'][3:7]
```

```python
s4g
# the difference between s4g and mi is that s4g has zero-based integer index, whereas mi has three indices (Symbol, 
# Year, and Month).
```

```python
mi
```

```python
test = s4g.groupby(['Symbol', 'Year', 'Month'], as_index=False)
test
# This output of aggregation is based on zero-based integer, rather than 3 indices ('Symbol', 'Year', and 'Month') 
# unlike the aggregation output generated above. The main difference is this aggregation is based on s4g, which has
# zero-based integer index, whereas the aggregation output generated above is based on mi, which has 3 indices ('Symbol,
# Year, and Month'). Apart from this index differences, the output of both aggregations is the same, that is, the same
# group means have been obtained in both cases (when the aggregation is based on s4g and on mi).
```

```python
type(test.agg(np.mean))
```

```python
test.agg(np.mean)
```

```python
mig_112.agg([np.mean, np.std])
# multiple statistics can be obtained throug a single aggregation call for all the gorups by passing a list of the 
# required statistics.
```
